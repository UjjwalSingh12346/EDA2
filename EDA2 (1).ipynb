```python
#2. Encoding Techniques
# Check all columns and the number of unique values in each column
print(data.columns)  # Print column names
```

```python
# Identify columns with less than 5 unique categories
categorical_columns = [col for col in data.select_dtypes(include=['object']).columns if data[col].nunique() < 5]
print("\nColumns with less than 5 unique categories:", categorical_columns)
```

```python
from sklearn.metrics import r2_score, accuracy_score
import pandas as pd
import numpy as np
```

```python
def calculate_pps(df, target):
    pps_matrix = pd.DataFrame(index=df.columns, columns=df.columns)
    
    for col1 in df.columns:
        for col2 in df.columns:
            if col1 == col2:
                pps_matrix.loc[col1, col2] = 1.0
            else:
                # Skip non-numeric columns
                if not np.issubdtype(df[col1].dtype, np.number) or not np.issubdtype(df[col2].dtype, np.number):
                    pps_matrix.loc[col1, col2] = np.nan
                    continue
                
                # Regression task (use r2_score for numeric columns)
                score = r2_score(df[col1], df[col2])
                pps_matrix.loc[col1, col2] = score
    
    return pps_matrix

# Example usage
pps_matrix = calculate_pps(data, target='target')
print(pps_matrix)
```
